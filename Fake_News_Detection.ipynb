```python
data.reset_index(inplace=True)
data.drop(['index'], axis=1, inplace=True)
```

```python
def wordopt(text):
  text = text.lower()
  text = re.sub('\[.*?\]', '', text)
  text = re.sub("\\W", " ", text)
  text = re.sub('https?://\S+|www\.\S+','',text)
  text = re.sub('<.*?>+', '', text)
  text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
  text = re.sub('\n', '', text)
  text = re.sub('\w*\d\w*', '', text)
  return text
```

```python
data['text'] = data['text'].apply(wordopt)
```

```python
x = data['text']
y = data['class']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size= 0.25,stratify=y, random_state=42)
```

```python
# Instead of fitting and transforming separately, fit on training data and transform both
vectorization = TfidfVectorizer()
xv_train = vectorization.fit_transform(x_train) # Fit on training data
xv_test = vectorization.transform(x_test) # Only transform the test data
```

```python
pred_lr = LR.predict(xv_test)
```

```python
def output_lable(n):
  if n == 0:
    return "Fake News"
  elif n == 1:
    return "Not a Fake News"
```

```python
def manual_testing(news):
  testing_news = {"text":[news]}
  new_def_test = pd.DataFrame(testing_news)
  new_def_test["text"] = new_def_test["text"].apply(wordopt)
  new_x_test = new_def_test["text"]
  new_xv_test = vectorization.transform(new_x_test)
  pred_LR = LR.predict(new_xv_test)
  pred_DT = DT.predict(new_xv_test)
  pred_GBC = GBC.predict(new_xv_test)
  pred_RFC = RFC.predict(new_xv_test)

  return print("\n\nLR Prediction: {} \nDT Prediction: {} \nGBC Prediction: {} \nRFC Prediction: {}".format(output_lable(pred_LR[0]), output_lable(pred_DT[0]), output_lable(pred_GBC[0]), output_lable(pred_RFC[0])))
```
